```python
import requests
from Bio import Entrez
Entrez.email = "raghavmangla04@gmail.com" 

# Initialize UniProt service
def get_uniprot_sequence(protein_name):
    """
    Retrieves the protein sequence from UniProt based on protein name.
    """
    try:
        # Construct the UniProt search URL
        url = f"https://www.uniprot.org/uniprotkb?query={protein_name}"
        response = requests.get(url)

        response.raise_for_status()  # Raise an exception for bad status codes
        
        # Extract the sequence from the JSON response
        data = response.json()
        print(data)
        if data:
            return data
        else:
            return None

    except requests.exceptions.RequestException as e:
        print(f"Error during UniProt request: {e}")
        return None
    
def get_ncbi_sequence(protein_name):
    """
    Retrieves the protein sequence from NCBI based on protein name.
    """
    try:
        # Search for the protein in NCBI
        handle = Entrez.esearch(db="protein", term=protein_name)
        record = Entrez.read(handle)
        handle.close()

        # Check if results were found
        if record['IdList']:
            # Fetch the sequence
            handle = Entrez.efetch(db="protein", id=record['IdList'][0], rettype="fasta", retmode="text")
            sequence = handle.read()
            handle.close()
            return sequence
        else:
            return None

    except Exception as e:
        print(f"Error during NCBI request: {e}")
        return None
```

```python
# Retrieve sequence from UniProt
uniprot_sequence = get_uniprot_sequence(protein_name)
if uniprot_sequence:
    print("UniProt Sequence:")
    print(uniprot_sequence)
else:
    print(f"Could not retrieve sequence from UniProt for {protein_name}")

# Retrieve sequence from NCBI
ncbi_sequences=[]
for protein in protein_names:
   ncbi_sequences.append(get_ncbi_sequence(protein))
if ncbi_sequences:
    print("\nNCBI Sequence:")
    
else:
    print(f"Could not retrieve sequence from NCBI for {protein_name}")
```

```python
cnt=0
protein_sequences=[]
for ncbi_sequence in ncbi_sequences:
    nc=""
    for i in range(len(ncbi_sequence)):
        if(ncbi_sequence[i]=='\n'):
            nc=ncbi_sequence[i+1:]
            break
        
    print(protein_names[cnt],'\n',nc)
    protein_sequences.append(nc)
    cnt=cnt+1

print(cnt)
```

```python
import torch
from transformers import EsmTokenizer, EsmModel

tokenizer = EsmTokenizer.from_pretrained("facebook/esm1b_t33_650M_UR50S")
model = EsmModel.from_pretrained("facebook/esm1b_t33_650M_UR50S")

# Generate Protein Embeddings
def get_protein_embedding(sequence):
    inputs = tokenizer(sequence, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze().numpy()

protein_embeddings = []

for seq in protein_sequences:
    protein_embeddings.append(get_protein_embedding(seq[0:1024]))
```

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client

# List of 50 protein names
protein_names = ["BRD4", "PD-1", "HER2", "EGFR", "BCR-ABL", "VEGFR", "CDK4/6", "PI3K (p110α)", "MTOR", "KRAS",
                 "BCL-2", "PARP1", "AKT1", "FLT3", "IDH1", "IDH2", "MAPT", "APP", "BACE1", "α-Synuclein",
                 "TDP-43", "HTT", "SOD1", "NMDA Receptor (GRIN2B)", "GSK-3β", "EPHX2", "ACE2", "PPARG", "PCSK9",
                 "HMGCR", "PPARα", "GLUT4 (SLC2A4)", "INSR", "AMPK (PRKAA1)", "GCK", "FABP4", "LEPR",
                 "HIV-1 Protease", "HIV-1 Reverse Transcriptase", "NS3/4A", "NS5B Polymerase", "SARS-CoV-2 Spike Protein",
                 "Mpro (3CLpro)", "CFTR", "NLRP3", "TRPV1", "FXR (NR1H4)", "P53 (TP53)", "KEAP1-NRF2", "mGluR5 (GRM5)"]

# ChEMBL client
target = new_client.target
activity = new_client.activity

all_data = []

# Iterate over each protein
for protein in protein_names:
    target_query = target.search(protein)
    targets = pd.DataFrame.from_dict(target_query)
    
    if not targets.empty:
        selected_target = targets.target_chembl_id[0]  # Take the first match
        
        # Get drug activity (IC50 values)
        res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
        df = pd.DataFrame.from_dict(res)

        if not df.empty:
            df = df.dropna(subset=['canonical_smiles', 'standard_value'])
            df['standard_value'] = pd.to_numeric(df['standard_value'], errors='coerce')
            df = df[df['standard_value'] > 0]

            df['protein'] = protein  # Add protein name column
            all_data.append(df[['molecule_chembl_id', 'canonical_smiles', 'standard_value', 'protein']])

# Combine all results
df_final = pd.concat(all_data, ignore_index=True)

# Save as CSV
df_final.to_csv("drug_protein_data.csv", index=False)
print("Data saved!")
```
